```python
def extract_text_from_pdf(pdf_path):
    text = ""

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"

    return text
```

```python
import nltk
import re
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return " ".join(tokens)
```

```python
sample = "I am a Software Developer skilled in Python, Machine Learning, and Data Science."
print(clean_text(sample))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def calculate_match_score(resume_text, job_text):

    documents = [resume_text, job_text]

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(documents)

    similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])

    return round(similarity[0][0] * 100, 2)
```

```python
resume_sample = """
Python developer with experience in machine learning, NLP and data science.
"""

job_sample = """
Looking for a machine learning engineer skilled in Python, NLP and AI.
"""

score = calculate_match_score(
    clean_text(resume_sample),
    clean_text(job_sample)
)

print("Matching Score:", score, "%")
```

```python
SKILLS_DB = [
    "python", "java", "c++", "machine learning", "deep learning", "data science",
    "nlp", "sql", "mysql", "mongodb", "html", "css", "javascript",
    "react", "node", "django", "flask", "aws", "docker", "kubernetes",
    "git", "github", "linux", "tensorflow", "pytorch", "opencv"
]
```

```python
def extract_skills(text):

    extracted = []

    for skill in SKILLS_DB:
        if skill in text:
            extracted.append(skill)

    return list(set(extracted))
```

```python
def find_missing_skills(resume_text, job_text):

    resume_skills = extract_skills(resume_text)
    job_skills = extract_skills(job_text)

    missing = list(set(job_skills) - set(resume_skills))

    return resume_skills, job_skills, missing
```

```python
resume = "python machine learning data science nlp"
job = "python sql docker aws machine learning nlp"

r_skills, j_skills, missing = find_missing_skills(
    clean_text(resume),
    clean_text(job)
)

print("Resume Skills:", r_skills)
print("Job Skills:", j_skills)
print("Missing Skills:", missing)
```
